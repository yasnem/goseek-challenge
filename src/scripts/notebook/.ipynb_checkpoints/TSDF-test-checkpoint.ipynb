```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib notebook
import matplotlib.pyplot as plt
from pathlib import Path
from tesse_gym.tasks.goseek import GoSeekFullPerception, decode_observations
from envs.utils import make_unity_env
from perception.tsdf_tools import TSDF 
import open3d
import numpy as np
```

```python
n_train_envs = 2
episode_length = 10
filename = Path("/home/jd/competition/goseek-challenge/simulator/goseek-v0.1.0.x86_64")
assert filename.exists(), f"Must set a valid path!"
```

```python
def visualize(rgb, segmentation, depth):
    ax[0].imshow(rgb[0])
    ax[1].imshow(segmentation[0])
    ax[2].imshow(depth[0])
```

```python
env = make_unity_env(filename, range(n_train_envs))
obs = env.reset()
rgb, segmentation, depth, pose = decode_observations(obs)

assert (
    n_train_envs % obs.shape[0] == 0
), f"The number of visualization environments must be a multiple of the training environments"
print("Finished loading environments.")
```

```python
from tqdm import tqdm

fig, ax = plt.subplots(1, 3, figsize=(18, 6))
ax[0].set_title('RGB')
ax[1].set_title('Semantics')
ax[2].set_title('Depth')

# Using a random agent.
action_space = np.arange(0, 4)
action_probability = np.array([0.3,0.3,0.3,0.1])
for i in tqdm(range(episode_length)):
    # give probability for actions in `action_space` front, left, right, pickup. 
    action = np.random.choice(action_space, p=action_probability)
    actions = np.zeros(obs.shape[0], dtype=np.int64)
    actions[0] = action
    obs, reward, done, pose = env.step(actions)

    plt.cla()
    rgb, segmentation, depth, pose = decode_observations(obs)
    visualize(rgb, segmentation, depth)
    fig.canvas.draw()
    if done[0]: 
        break
        
```
